# **Data Preprocesing**

## **Feature Engineering**

```python
dataset = pd.read_csv('/content/drive/MyDrive/Data Science/Flight Fare Prediction/Data_Train.csv')
dataset.head()
```

```python
dataset['Journey_Day'] = pd.to_datetime(dataset.Date_of_Journey, format='%d/%m/%Y').dt.day
dataset['Journey_Month'] = pd.to_datetime(dataset.Date_of_Journey, format='%d/%m/%Y').dt.month
```

```python
dataset.head()
```

```python
dataset['Dep_Hour'] = pd.to_datetime(dataset.Dep_Time).dt.hour
dataset['Dep_Min'] = pd.to_datetime(dataset.Dep_Time).dt.minute
dataset.head()
```

```python
dataset['Arriva_Hour'] = pd.to_datetime(dataset.Arrival_Time).dt.hour
dataset['Arrival_Min'] = pd.to_datetime(dataset.Arrival_Time).dt.minute
dataset.head()
```

```python
dataset.drop(columns=['Date_of_Journey', 'Route', 'Dep_Time', 'Arrival_Time', 'Additional_Info'], inplace=True)
dataset.head()
```

```python
# Splitting Duration Column
# Initialize new columns
dataset['Duration_hour'] = np.nan
dataset['Duration_minute'] = np.nan

# Extract hours and minutes using regex
dataset['Duration_hour'] = dataset['Duration'].str.extract(r'(\d+)h').astype(float)  # Extract hours
dataset['Duration_minute'] = dataset['Duration'].str.extract(r'(\d+)m').astype(float)  # Extract minutes

# Fill NaN values with 0 for cases where there are no hours or minutes
dataset['Duration_hour'] = dataset['Duration_hour'].fillna(0).astype(int)
dataset['Duration_minute'] = dataset['Duration_minute'].fillna(0).astype(int)
```

```python
dataset.head()
```

```python
Airline = dataset['Airline']
Airline = pd.get_dummies(Airline, drop_first=True).astype(int)
Airline.head()
```

```python
Source = dataset['Source']
Source = pd.get_dummies(Source).astype(int)
Source.head()
```

```python
Destination = dataset['Destination']
Destination = pd.get_dummies(Destination).astype(int)
Destination.head()
```

```python
dataset.head()
```

```python
dataset = pd.concat([dataset, Airline, Source, Destination], axis=1)
dataset.head()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
# Convert Total_Stops to integer using lambda
dataset['Total_Stops'] = dataset['Total_Stops'].apply(lambda x: 0 if x == 'non-stop' else int(x.split()[0]))
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
dataset.head()
```

```python
dataset.drop(columns=['Airline', 'Source', 'Destination', 'Duration'], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
Price = dataset['Price']
Price.head()
```

```python
dataset.drop('Price', axis=1, inplace=True)
```

```python
dataset = pd.concat([dataset, Price], axis=1)
dataset.head()
```

## **Splitting the dataset**

```python
# Split the data into X and y
X = dataset.iloc[:, :-1].values  # All columns except the last one
y = dataset.iloc[:, -1].values   # Only the last column
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# **Making an ANN**

```python
# Importing the Keras
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initializing the ANN
model = Sequential()
```

```python
# Adding the input layer and the first hidden layer
model.add(Dense(units=16, kernel_initializer='uniform', activation='relu', input_dim = X.shape[1]))
```

```python
# Adding additional hidden layers
model.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
model.add(Dense(units=16, kernel_initializer='uniform', activation='relu'))
```

```python
# Adding the output layer
model.add(Dense(units=1, kernel_initializer='uniform', activation='linear'))
```

```python
# Compiling the ANN
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
```

```python
# fitting the ANN to the Training set
history = model.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Calculating MAE
mae = mean_absolute_error(y_test, y_preds)
print(f'Mean Absolute Error: {mae}')

# Calculating MSE
mse = mean_squared_error(y_test, y_preds)
print(f'Mean Squared Error: {mse}')

# R2_Score
r2 = r2_score(y_test, y_preds)
print(f'R2 Score: {r2}')

# Calculating RMSE
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```

# **Matplotlib Example Workflow**

```python
# 0. import matplotlib and get  it ready  for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1.  Prepare Data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
# 5. Save Figure
fig.savefig("sample-plot.png")
```

```python
# Crete a Plot from a Dicttionary

nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Nut Butter Store",
       ylabel="Price ($)",);
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

# **Two options for Subplots**

```python
# Option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5));
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2
fig, ax = plt.subplots(nrows=2,
                 ncols=2,
                 figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("/content/heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease['age'].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# PyPlot Method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO Method mixed with PyPlot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax = ax );
```

```python
# OO Method from Scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the Graph
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
# Name the Axes
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a Mean Line of Cholesterol
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize=(10,6))

sc = ax.scatter(x=over_50['age'],
                y=over_50['thalach'],
                c=over_50['target'],
                cmap='winter')

ax.set(title="Heart Disease and Heart Rate",
       xlabel='Age',
       ylabel='Heart Rate/ Thalach')

ax.legend(*sc.legend_elements(), title="Target")

ax.axhline(over_50['thalach'].mean(),
           linestyle='--');
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# Add Data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

# Add Data to ax2
scatter1 = ax2.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])

# Name the Axes of ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Name the Axes of ax2
ax2.set(title="Heart Disease and  Max Heart Rate",
       xlabel='Age',
       ylabel=' Max Heart Rate/ Thalach')

# Add a Legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a Legend to ax2
ax2.legend(*scatter.legend_elements(), title="Target")

# Add a Mean Line of Cholesterol
ax1.axhline(over_50['chol'].mean(),
           linestyle='--')

# Add a Mean Line of Thalach
ax2.axhline(over_50['thalach'].mean(),
           linestyle='--')

# Add a title to the Figure
fig.suptitle("Heart Disease Ananlysis", fontsize=16, fontweight='bold');
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# Add Data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')

# Add Data to ax2
scatter1 = ax2.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter')

# Change the x and y axis limits of ax1
ax1.set_xlim([50,80])

# Change the x and y axis limits of ax2
ax2.set_xlim([50,80])
ax2.set_ylim([60,200])

# Name the Axes of ax1
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Name the Axes of ax2
ax2.set(title="Heart Disease and  Max Heart Rate",
       xlabel='Age',
       ylabel=' Max Heart Rate/ Thalach')

# Add a Legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a Legend to ax2
ax2.legend(*scatter.legend_elements(), title="Target")

# Add a Mean Line of Cholesterol
ax1.axhline(over_50['chol'].mean(),
           linestyle='--',
            color='red')

# Add a Mean Line of Thalach
ax2.axhline(over_50['thalach'].mean(),
           linestyle='--',
            color='red')

# Add a title to the Figure
fig.suptitle("Heart Disease Ananlysis", fontsize=16, fontweight='bold');
```
